```python
import glob
import os.path
import pandas as pd
import math
from datetime import datetime, timedelta
```

```python
pa_dir="V:/ACOI/R01 - W4K/4_Free living/Participant Data/"
participants=glob.glob(pa_dir+"[0-9][0-9][0-9][0-9]/")
tracking_sheet = pd.read_excel("V:/ACOI/R01 - W4K/4_Free living/Free-living master tracking.xlsx",skiprows=1)
```

```python
cur_par=1
for participant in participants:
    part_id=participant[len(pa_dir):len(pa_dir)+4]
    
    print(f"Participant {part_id} {cur_par}/{len(participants)}")
    print("Camp Data")
    try:
        comparetimes(part_id,"Camp")
    except Exception as error:
        # handle the exception
        print("An exception occurred:", error)
    print("Home Data")
    try:
        comparetimes(part_id,"Home")
    except Exception as error:
        # handle the exception
        print("An exception occurred:", error)
    

            
        
    
    cur_par+=1
```

```python
def comparetimes(part_id,protocol):
    if protocol=="Camp":      
        start_time = tracking_sheet.loc[tracking_sheet['WDID'] == int(part_id), 'ALL DEVICES ON: CAMP'].iloc[0]
        start_time = start_time.strftime("%I:%M %p")  # Format in 12-hour clock with AM/PM
        end_time = tracking_sheet.loc[tracking_sheet['WDID'] == int(part_id), 'ALL DEVICES OFF: CAMP'].iloc[0]
        end_time = end_time.strftime("%I:%M %p")  # Format in 12-hour clock with AM/PM
        # Read CSV file
        protocol=protocol.lower()
        st_time = pd.read_csv(pa_dir+f"{part_id}/{protocol}/{part_id}_{protocol}_log.csv")
        st_time_datetime = pd.to_datetime(st_time.loc[0, 'Activity_Start'])  # Convert to datetime object
        formatted_st_time = st_time_datetime.strftime('%m/%d/%Y %I:%M %p')  # Format as desired
        if formatted_st_time[-8:] == start_time:
            print("Start time matches")
        else:
            print("Start Time does not Match")
            tracking_sheet.loc[tracking_sheet['WDID'] == int(part_id), 'ALL DEVICES ON: CAMP'].iloc[0]=formatted_st_time[-8:]
        ed_time_datetime = pd.to_datetime(st_time.iloc[-1, 3])
        formatted_ed_time = ed_time_datetime.strftime('%m/%d/%Y %I:%M %p')
        if formatted_ed_time[-8:] == end_time:
            print("End time matches")
        else:
            print("End time does not match")
            
    if protocol=="Home":
        start_time = tracking_sheet.loc[tracking_sheet['WDID'] == int(part_id), 'ALL DEVICES ON: HOME'].iloc[0]
        start_time = start_time.strftime("%I:%M %p")  # Format in 12-hour clock with AM/PM
        end_time = tracking_sheet.loc[tracking_sheet['WDID'] == int(part_id), 'ALL DEVICES OFF: HOME'].iloc[0]
        end_time = end_time.strftime("%I:%M %p")  # Format in 12-hour clock with AM/PM
        # Read CSV file
        protocol=protocol.lower()
        st_time = pd.read_csv(pa_dir+f"{part_id}/{protocol}/{part_id}_{protocol}_log.csv")
        st_time_datetime = pd.to_datetime(st_time.iloc[0,0])  # Convert to datetime object
        formatted_st_time = st_time_datetime.strftime('%m/%d/%Y %I:%M %p')  # Format as desired
        if formatted_st_time[-8:] == start_time:
            print("Start time matches")
        else:
            print("Start Time does not Match")
        
        ed_time_datetime = pd.to_datetime(st_time.iloc[-1, 0])
        formatted_ed_time = ed_time_datetime.strftime('%m/%d/%Y %I:%M %p')
        
        if formatted_ed_time[-8:] == end_time:
            print("End time matches")
        else:
            print("End time does not match")
        
```
