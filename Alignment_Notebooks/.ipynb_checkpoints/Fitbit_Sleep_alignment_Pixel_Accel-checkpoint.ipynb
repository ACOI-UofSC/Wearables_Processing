```python
def timestamp_pixel(accel_file, participant_id):
    # Read in accel data while dropping first column,
    accel_list = []
    for file in accel_file:
        accel_list.append(pd.read_csv(file, header=None, names=['Counter', 'Time', 'X', 'Y', 'Z'], usecols=[1, 2, 3, 4],
                                      parse_dates=[0], date_parser=lambda x: datetime.fromtimestamp(int(x) / 1000)))
    if len(accel_list) > 1:
        accel_data = pd.concat(accel_list, ignore_index=True)
    else:
        accel_data = accel_list[0]
    accel_data[['X', 'Y', 'Z']] = accel_data[['X', 'Y', 'Z']].apply(pd.to_numeric)
    accel_data["Pixel ENMO"]=calc_enmo(accel_data)
    print(accel_data)
    return accel_data
```

```python
def calc_enmo(some_data):
    # Convert X, Y, Z columns to numeric
    
    # Drop rows with NaN values
    some_data = some_data.dropna(subset=['X', 'Y', 'Z'])
    
    # Calculate the magnitude by first squaring all of the x, y, and z value, then summing them, and taking the square root.
    mag = ((some_data[['X', 'Y', 'Z']].applymap(lambda x: x ** 2)).sum(axis=1)).transform(lambda x: np.sqrt(x))
    # To calculate ENMO we subtract 1 (Gravity) from the vector magnitudes
    enmo = mag.transform(lambda x: ((x/9.8)-1))
    # Finally, if we have any ENMO values less than 0, we round them up.
    enmo.loc[enmo.loc[:] < 0] = 0

    return enmo
```
