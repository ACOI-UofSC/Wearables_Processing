```python
# Read in tracking sheet to obtain protocol date
tracking_sheet = pd.read_excel("V:\\ACOI\\R01 - W4K\\3_PA protocol\\PA master tracking.xlsx").dropna(how='all')
date = tracking_sheet.loc[tracking_sheet['WDID'] == int(part_id), 'PROTOCOL DATE'].iloc[0]
# Read in activity log to obtain protocl start and end
activity_log = pd.read_excel(path + "Survey and Protocol documents\\Activity time log.xlsx", usecols=['Activity', 'Start time', 'End time'])
start = datetime.combine(date.date(), activity_log.iloc[0,1])
end = datetime.combine(date.date(), activity_log.iloc[-1,2])
print(f"Protocol Start: {start} \nProtocol End: {end}")
```

```python
# Read in Garmin data
slog_hr = pd.read_csv(path + "Garmin data\\" + part_id + "_1_raw_data.csv", parse_dates=['record.timestamp[s]'], infer_datetime_format=True)
# Read in FITABASE data
#fitabase = pd.read_csv(path + "Fitbit\\Fitabase\\" + part_id + "_hr.csv", parse_dates=['Time'])

# Assuming 'record.timestamp[s]' is currently in Unix timestamp format
slog_hr['record.timestamp[s]'] = pd.to_datetime(slog_hr['record.timestamp[s]'], unit='s', errors='coerce')

# Add a timedelta to the 'record.timestamp[s]' column
slog_hr['record.timestamp[s]'] = slog_hr['record.timestamp[s]'] - pd.Timedelta(days=0, hours=0, minutes=3,seconds=15)

# Assuming 'record.timestamp[s]' is currently in datetime format
slog_hr['record.timestamp[s]'] = (slog_hr['record.timestamp[s]'] - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')


print(slog_hr['record.timestamp[s]'])
```

```python
garmin_enmo=enmo = pd.read_csv(path + part_id + "_agg.csv", parse_dates=['Time'], infer_datetime_format=True)
acti_enmo=enmo = pd.read_csv(path + part_id + "_agg.csv", parse_dates=['Time'], infer_datetime_format=True)
garmin_enmo["Time"]=garmin_enmo["Time"].apply(lambda x: x - timedelta(hours=0,minutes=3,seconds=15))

# Plot HR to check initial alignment
fig, ax = plt.subplots(figsize=[25,15])
for device in acti_enmo.columns:
    if "Actigraph Max ENMO" in device:
        temp = acti_enmo[["Time", device]].dropna()
        ax.plot(temp["Time"], temp[device], label="Actigraph", color="blue")
    
for device in garmin_enmo.columns:
    if "Garmin Max ENMO" in device:
        temp = garmin_enmo[["Time", device]].dropna()
        ax.plot(temp["Time"], temp[device], label="Garmin", color="green")
#ax.plot(actiheart['Time'], actiheart['Heart Rate'], label='Actiheart', color='blue')
ax.set(xlim=[start, end], ylim=[0,10])
fig.legend(fontsize='xx-large')
plt.show()
```
